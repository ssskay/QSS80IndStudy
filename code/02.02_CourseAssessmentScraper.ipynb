```python
# builds course assessment table
import time
course_review_table_list = []

for index, row in course_assess_df.iterrows():
    # goes to course assessment link
    driver.get(row['Link'])
    
    # provides buffer time for table info to load
    time.sleep(4)
    try:
        # gets review tables
        tables = driver.find_elements_by_class_name("PTChildPivotTable")[2:]
        reviews_list = []

        # cleans table
        for t in tables:
            review_table = pd.read_html(t.get_attribute('outerHTML'))[0]
            review_table = review_table.dropna()
            review_table = review_table.rename(columns={0: "Term", 1:'Review'})
            review_table = review_table[review_table['Review'] != 'Comments']
            reviews_list.append(review_table)

        # connects info from course assessment directory to df
        if len(reviews_list) > 0:
            course_comments = pd.concat(reviews_list)
            course_comments['Department'] = row['Department']
            course_comments['Course'] = row['Course Name']
            course_comments['Prof'] = row['Prof Name']
            course_review_table_list.append(course_comments)
    except:
        continue
    print('Scraped ' + row['Course Name'])
```

```python
# merges all courses into one df
courseassessmentreviews = pd.concat(course_review_table_list)
courseassessmentreviews
```

```python
courseassessmentreviews.to_csv ('../data/course_assessment_reviews.csv', index = False, header=True)
```
